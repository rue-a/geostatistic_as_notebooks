Berechnen Sie die statistischen Lage- und Streuungsmaße Median, Mittelwert, Spannweite, Interquartilsabstand und Standardabweichung für die Lufttemperatur für die DWD-Station in Dresden. Zeichnen Sie mit Hilfe von R einen Boxplot. Der Boxplot soll auch das arithmetische Mittel als Kreis mit anzeigen und beschriftet sein.

```r
dwdsr <- read.csv2(file = "Data_Beleg3.csv", header = TRUE)

# transform to numeric:
dwdsr$TT_TU <- as.numeric(dwdsr$TT_TU)

# median: 
print(paste("Der Median beträgt", median(dwdsr$TT_TU)))

# Mean:     
print(paste("Der Mittelwert beträgt", mean(dwdsr$TT_TU)))

# Spannweite:    
print(paste("Die Spannweite beträgt", max(dwdsr$TT_TU) - min(dwdsr$TT_TU)))
# oder:
print("Die Werte liegen in folgedem Bereich:")
range(dwdsr$TT_TU)

# Interquantilsabstand: 
print(paste("Der Interquantilsabstand beträgt", quantile(dwdsr$TT_TU, 0.75) - quantile(dwdsr$TT_TU, 0.25)))

# Standardabweichung: 
print(paste("Die Standardabweichung beträgt", sd(dwdsr$TT_TU)))

# boxplot: 
boxplot(dwdsr$TT_TU, main = "Boxplot der Lufttemperatur ", ylab = "Lufttemperatur in [°C]")
points(mean(dwdsr$TT_TU),col="blue")

# Schiefegrad: 
skewness(dwdsr$TT_TU)
```
